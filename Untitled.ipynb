```python
# Filtramos valores anómalos: eliminamos edades mayores a 90 (ajustable según criterio)
edad_filtrada = tabla[(tabla['EDAD_PADRE'] >= 13) & (tabla['EDAD_PADRE'] <= 90)]['EDAD_PADRE']

# Graficamos el histograma con KDE
plt.figure(figsize=(10, 6))
sns.histplot(edad_filtrada.dropna(), bins=30, kde=True, color='lightcoral')
plt.title('Histograma de EDAD_PADRE con estimación KDE')
plt.xlabel('Edad del padre')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm
import numpy as np

# Datos limpios
edad_filtrada = tabla[(tabla['EDAD_PADRE'] >= 13) & (tabla['EDAD_PADRE'] <= 90)]['EDAD_PADRE'].dropna()

# Media y desviación estándar de los datos
mu = edad_filtrada.mean()
sigma = edad_filtrada.std()

# Crear rango de valores para la curva normal
x = np.linspace(edad_filtrada.min(), edad_filtrada.max(), 100)

# Densidad de probabilidad de la normal
pdf = norm.pdf(x, mu, sigma)

# Gráfico
plt.figure(figsize=(10, 6))
sns.histplot(edad_filtrada, bins=30, kde=True, stat='density', color='skyblue', label='KDE real')
plt.plot(x, pdf, 'r--', label='Distribución normal teórica')
plt.title('EDAD_PADRE: KDE vs Distribución Normal')
plt.xlabel('Edad del padre')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```

Comparación de la variable EDAD_PADRE con la distribución normal
Se construyó un histograma con estimación de densidad (KDE) para la variable EDAD_PADRE, y se superpuso una curva de distribución normal teórica usando la media y desviación estándar de los datos filtrados.

 Observaciones:
La curva KDE azul representa la distribución real de las edades de los padres (entre 13 y 90 años).

La línea roja punteada representa una distribución normal con la misma media y desviación estándar de los datos.

 Análisis:
La distribución real es asimétrica hacia la derecha (sesgo positivo), lo que significa que hay una mayor concentración de padres jóvenes, pero algunos casos menos frecuentes de padres mayores.

Aunque la media es de aproximadamente 31 años, la forma no es perfectamente simétrica como la distribución normal.

Esto indica que no se cumple estrictamente el supuesto de normalidad, por lo que es conveniente usar pruebas estadísticas para confirmar esta observación (por ejemplo, Shapiro-Wilk).

```python
from scipy.stats import shapiro

# Aplicar la prueba de Shapiro-Wilk a EDAD_PADRE
edad_padre = tabla['EDAD_PADRE'].dropna()

# Realizar la prueba de normalidad
stat, p_value = shapiro(edad_padre)

# Mostrar resultados
print(f'Estadístico de Shapiro-Wilk: {stat}')
print(f'Valor-p: {p_value}')

# Interpretación
if p_value > 0.05:
    print("No se rechaza la hipótesis nula: los datos parecen seguir una distribución normal.")
else:
    print("Se rechaza la hipótesis nula: los datos no siguen una distribución normal.")
```

¨El test de Shapiro-Wilk es un metodo calsico y ampliamnete aceptado para verificar si una variable sigue una distribucion normal¨ (Shapiro & wilk , 1965)

Realizamos la prueba de normalidad de Shapiro-Wilk sobre los datos de `EDAD_PADRE`:


- Estadístico de Shapiro-Wilk**: (valor calculado).
- Valor-p: (valor-p calculado).

Interpretación:

- Si el valor-p es mayor que 0.05, no se rechaza la hipótesis nula y se considera que los datos siguen una distribución normal.
- Si el valor-p es menor o igual a 0.05, se rechaza la hipótesis nula y los datos no siguen una distribución normal.

En nuestro caso, el valor-p es (valor-p calculado), por lo tanto:

- Si es mayor a 0.05, los datos parecen seguir una distribución normal.
- Si es menor a 0.05, se rechaza la hipótesis de normalidad.

Este análisis respalda las observaciones realizadas en el gráfico de densidad.

```python
from scipy.stats import pearsonr, spearmanr, kendalltau

# Guardamos las columnas en variables
edad_padre = tabla_limpia['EDAD_PADRE']
n_emb = tabla_limpia['N_EMB']

# Pearson
pearson_corr, _ = pearsonr(edad_padre, n_emb)

# Spearman
spearman_corr, _ = spearmanr(edad_padre, n_emb)

# Kendall
kendall_corr, _ = kendalltau(edad_padre, n_emb)

# Mostramos los resultados
print(f'Coeficiente de Pearson: {pearson_corr:.4f}')
print(f'Coeficiente de Spearman: {spearman_corr:.4f}')
print(f'Coeficiente de Kendall: {kendall_corr:.4f}')
```

Coeficiente de Kendall: 0.2962

Estos valores indican una correlación moderada y positiva entre la edad del padre y el número de embarazos. Es decir, a medida que aumenta la edad del padre, también tiende a aumentar el número de embarazos registrados.

 La correlación de Pearson mide una relación lineal, mientras que Spearman y Kendall analizan relaciones monótonas. Los tres coeficientes están relativamente cercanos, lo cual indica una tendencia consistente y ascendente entre las variables, aunque no necesariamente perfectamente lineal.
